```python
train=pd.read_csv("/kaggle/input/titanic/train.csv")
test=pd.read_csv("/kaggle/input/titanic/test.csv")
print(train.head())

```

```python
print(test.head())
```

```python
y_train=train.Survived
x_train=train.drop('Survived',axis=1)

features=['Pclass','Sex','Age','SibSp','Fare']
x_train=x_train[features]
test=test[features]
print(x_train.head())
print(test.head())
```

```python
x_train.Sex[x_train.Sex == 'male'] = 1
x_train.Sex[x_train.Sex == 'female'] = 2
test.Sex[test.Sex == 'male'] = 1
test.Sex[test.Sex == 'female'] = 2
print(x_train.head())
print(test.head())
x_train=x_train.fillna(0)
test=test.fillna(0)

pd.isnull(test)
```

```python
x_train=x_train.values

test=test.values

```

```python

print(test)
```

```python
x_train,x_val,y_train,y_val=train_test_split(x_train,y_train,test_size=0.2,shuffle=True)
x_train=x_train.astype('float32')
y_train=y_train.astype('float32')
x_val=x_val.astype('float32')
y_val=y_val.astype('float32')
test=test.astype('float32')

tf.convert_to_tensor(x_train)
tf.convert_to_tensor(y_train)
tf.convert_to_tensor(x_val)
tf.convert_to_tensor(y_val)
tf.convert_to_tensor(test)

```

```python
print(x_train.shape,x_val.shape)
```

```python
model = tf.keras.Sequential()

batch_size = 16
epochs = 100

model.add(tf.keras.layers.Dense(64, activation="relu",kernel_initializer=tf.keras.initializers.GlorotNormal(seed=1), input_shape=(None,5)))
model.add(tf.keras.layers.Dense(32, activation="relu",kernel_regularizer=tf.keras.regularizers.l1_l2(l1=1e-5, l2=1e-4)))
model.add(tf.keras.layers.Dropout(0.25))
model.add(tf.keras.layers.Dense(16, activation="tanh",kernel_initializer=tf.keras.initializers.GlorotNormal(seed=1)))
model.add(tf.keras.layers.Dropout(0.25))
model.add(tf.keras.layers.Dense(8, activation="tanh",kernel_initializer=tf.keras.initializers.GlorotNormal(seed=1)))
model.add(tf.keras.layers.Dropout(0.25))
model.add(tf.keras.layers.Dense(1, activation="sigmoid"))




model.compile(optimizer=tf.keras.optimizers.RMSprop(lr=0.001,clipvalue=5),loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history=model.fit(x_train, y_train, epochs=epochs, batch_size=batch_size, validation_data=(x_val,y_val))
```

```python
y=model.predict(test)


for i in range(0,len(y)):
    if y[i]<0.5:
        y[i]=0
    else:
        y[i]=1
print(y)
```
